```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_pi(num_samples):
    # Generate random points
    x = np.random.uniform(0, 1, num_samples)
    y = np.random.uniform(0, 1, num_samples)

    # Calculate the distance from the origin
    distance = np.sqrt(x**2 + y**2)

    # Count points inside the quarter circle
    inside_circle = distance <= 1

    # Estimate π
    pi_estimate = 4 * np.sum(inside_circle) / num_samples
    
    return pi_estimate

# Number of samples
num_samples = 1000000

# Run the Monte Carlo simulation
pi_estimate = monte_carlo_pi(num_samples)

# Print the result
print(f"Estimated value of π: {pi_estimate}")

# Optional: Visualize the simulation
x = np.random.uniform(0, 1, num_samples)
y = np.random.uniform(0, 1, num_samples)
inside_circle = x**2 + y**2 <= 1

plt.figure(figsize=(8,8))
plt.scatter(x[inside_circle], y[inside_circle], color='yellow', alpha=0.5, label='Inside Circle')
plt.scatter(x[~inside_circle], y[~inside_circle], color='violet', alpha=0.5, label='Outside Circle')
plt.legend()
plt.title('Monte Carlo Simulation of π')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
